## Part 1 - Word Counts  

**Read in the text file** - In the cell below, read the *gettysburg_address.txt* file (on Canvas) into Python.

```python
f= open("gettysburg_address.txt","r")
data = str(f.read())
print(data)
```

**Question:** What is the difference between **read()** and **readlines()**?

read() presents it how you would see in a text file where as readlines shows how you would type it by including the \n and such to seperate the lines.

**Split the text** - Split the text into a list of individual words. This process is called tokenization and is a common data pre-processing step when working with any text data.

```python
tokenizedList = data.split()
```

**Create a dictionary** - Initialize an empty dictionary to hold the word counts.

```python
Dict = {}
```

**Populate your dictionary** - Iterate over your list of words and update the word counts for each word in the text.

```python
for word in tokenizedList:
    if word not in Dict:
        Dict[word] = 0
    Dict[word] += 1
    
if '-' in Dict:
    del Dict['-']

print(Dict)
```

**Sort your word count dictionary** - Dictionaries are unordered collections in Python. You will need to find a way to sort the dictionary by its values. Save the sorted output to a new variable.

Example output:  
Dictionary: {'of': 14, 'was': 5, 'The': 27}  
Sorted output: ('The', 27), ('of', 14), ('was', 5)

```python
import operator
sortedDict = sorted(Dict.items(), key=operator.itemgetter(1), reverse=True)

print(sortedDict)
```

**Question:** How does **items()** return the dictionary key-value pairs?

.items() returns the dictionary key and values as tuples instead of listed out with semicolon seperation

**Print out the top 10 most used words.**

```python
for k in range (0,10):
    print(list(sortedDict)[k])
```

<b>Import the dataset</b> - Use the cell below to import the *voting_data.csv* file (on Canvas) and load in the data. This dataset shows how a set of congress-people voted on various issues. 

```python
import pandas as pd

data = pd.read_csv('voting_data.csv')
```

<b>Change a column name</b> - Retitle the column labelled 'Class Name' with 'political_party'.

```python
data.rename(columns={'Class Name':'political_party'}, inplace=True)
data.head()
```

<b>Replacing values</b> - In most cases, it is easier to work with integers rather than other data types. Whenever possible, it is suggested that you convert characters and strings to integer values. Use the following cell to replace 'n' with 0 and 'y' with 1 in each row.

```python
data.replace(to_replace=['n','y'], value=[0,1], inplace=True)
data.head()
```

<b>Count party members</b> - Print out the number of republicans and democrats in this congress .

```python
print(data['political_party'].value_counts())
```

<b>Which votes pass?</b> - Determine which issues would pass (more yes's than no's) and which issues would not pass. Print your results.

```python
for col in data.columns:
    if col is 'political_party':
        continue;
    voteCount = data[col].value_counts(sort=True) #count the columns 1 and 0s then put the more frequent one first
    #print(list(voteCount.to_dict().keys())[0])
    if(list(voteCount.to_dict().keys())[0] == 1):#if more people voted 1 (yes) then it passed
        print(col,'vote passed')
    if(list(voteCount.to_dict().keys())[0] == 0):#if more people voted 0 (no) then it failed
        print(col,'vote failed')
```
